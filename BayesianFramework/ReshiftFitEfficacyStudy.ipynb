```python
def genGroupedRedshiftFitSubplots(data, groupbyCol, redshifts, groupByColUnit = None, restrictToModelledRedshifts = False):
    redshiftRange = (0, 5) if not restrictToModelledRedshifts else (np.amin(np.unique(np.concatenate(redshifts))) - 0.1, np.amax(np.unique(np.concatenate(redshifts)))+0.1)
    supTitlePrefix = '' if not restrictToModelledRedshifts else 'Redshift range restricted to modelled values. '
    grouped = data.groupby(groupbyCol)
    numGroups = grouped.ngroups
    # make subplots with three plots per row and as many rows as required
    numRows = np.ceil(numGroups/3)
    figure, axes = mplplot.subplots(figsize=(9,3*numRows), 
                                    nrows=int(numRows), 
                                    ncols=3)
    for plotIndex, (groupKey, plotAxis) in enumerate(zip(sorted(grouped.groups.keys()), axes.flatten())) :
        groupData = grouped.get_group(groupKey)
        roiData = groupData[groupData.ROI.astype(bool)]
        contamData = groupData[~groupData.ROI.astype(bool)]
        if len(roiData.index) > 0 :
            roiData.BEST_FIT_REDSHIFT.plot.hist(ax=plotAxis, histtype='step', fill=True, ec=redshiftSubsetColours['ROI'], fc=redshiftSubsetColours['ROI'], alpha=0.3, bins=50, range=redshiftRange, label='ROI')
        if len(contamData.index) > 0 :
            contamData.BEST_FIT_REDSHIFT.plot.hist(ax=plotAxis, histtype='step', fill=True, ec=redshiftSubsetColours['CONTAM'], fc=redshiftSubsetColours['CONTAM'], alpha=0.3, bins=50, range=redshiftRange, label='CONTAM')
        plotAxis.set_title('{} = {:4.2f} {}'.format(groupbyCol, groupKey, groupByColUnit if groupByColUnit is not None else ''))
        plotAxis.set_xlabel('Best Fit Redshift')
        vlinesYmin, vlinesYmax = plotAxis.get_ylim()[0], plotAxis.get_ylim()[1]
        for redshiftSubset, colour in zip(redshifts, [redshiftSubsetColours['ROI'], redshiftSubsetColours['CONTAM']]) :
            plotAxis.vlines(redshiftSubset, ymin=vlinesYmin, ymax=vlinesYmax, colors=colour, alpha=0.1)
        plotAxis.legend()
            
    mplplot.suptitle(supTitlePrefix + groupbyCol, y=1.01)
    mplplot.tight_layout()
    
def genGroupedRedshiftFitDeltaSubplots(data, groupbyCol, redshifts, groupByColUnit = None):
    grouped = data.groupby(groupbyCol)
    numGroups = grouped.ngroups
    # make subplots with three plots per row and as many rows as required
    numRows = np.ceil(numGroups/3)
    figure, axes = mplplot.subplots(figsize=(9,3*numRows), 
                                    nrows=int(numRows), 
                                    ncols=3)
    for plotIndex, (groupKey, plotAxis) in enumerate(zip(sorted(grouped.groups.keys()), axes.flatten())) :
        groupData = grouped.get_group(groupKey)
        roiData = groupData[groupData.ROI.astype(bool)]
        contamData = groupData[~groupData.ROI.astype(bool)]
        if len(roiData.index) > 0 :
            roiData.DELTA.plot.hist(ax=plotAxis, histtype='step', fill=True, ec=redshiftSubsetColours['ROI'], fc=redshiftSubsetColours['ROI'], alpha=0.3, bins=50, label='ROI')
        if len(contamData.index) > 0 :
            contamData.DELTA.plot.hist(ax=plotAxis, histtype='step', fill=True, ec=redshiftSubsetColours['CONTAM'], fc=redshiftSubsetColours['CONTAM'], alpha=0.3, bins=50, label='CONTAM')
        plotAxis.set_title('{} = {:4.2f} {}'.format(groupbyCol, groupKey, groupByColUnit if groupByColUnit is not None else ''))
        plotAxis.set_xlabel(r'$\frac{z_{\rm fit} - z_{\rm true}}{1+z_{\rm true}}$', size=15)
        plotAxis.legend()
        
    mplplot.suptitle(groupbyCol, y=1.01)
    mplplot.tight_layout()
```

# Categorized plots of difference ($\frac{z_{\rm fit} - z_{\rm true}}{1+z_{\rm true}}$) versus $z_{\rm true}$ and $z_{\rm true}$ 

```python
def genCategorizedZVersusDeltaPlots(data, categoryColumn, categoryLabelFunctor, colorMapName='viridis') :
    figure, deltaStatsAxes = mplplot.subplots(figsize=(16, 8),
                                    ncols=2,
                                    nrows=1)

    groupedData = data.groupby(categoryColumn)
    colourMap = matplotlib.cm.get_cmap(colorMapName)
    indices = np.unique(data[categoryColumn].values)
    minIndex, maxIndex = np.amax(indices), np.amin(indices)
    groupColours = {int(index) : colourMap((index-minIndex)/(maxIndex - minIndex)) for index in indices} 

    for groupIndex, (key, group) in enumerate(groupedData):
        group.plot.scatter(ax=deltaStatsAxes[0], x='DELTA', y='BEST_FIT_REDSHIFT', logx=False, label=categoryLabelFunctor(key), s=100, alpha=0.5, color=groupColours[int(key)])
        group.plot.scatter(ax=deltaStatsAxes[1], x='DELTA', y='REDSHIFT', logx=False, label=categoryLabelFunctor(key), s=100, alpha=0.5, color=groupColours[int(key)])

    deltaStatsAxes[0].set_xlabel(r'$\frac{z_{\rm fit} - z_{\rm true}}{1+z_{\rm true}}$', size=20)    
    deltaStatsAxes[1].set_xlabel(r'$\frac{z_{\rm fit} - z_{\rm true}}{1+z_{\rm true}}$', size=20) 

    deltaStatsAxes[0].set_ylabel(r'$z_{\rm fit}$', size=20)    
    deltaStatsAxes[1].set_ylabel(r'$z_{\rm true}$', size=20)  
    
    mplplot.tight_layout()
    return deltaStatsAxes
```

# Compute contamination fraction
This is the fraction of spectra with true (modelled) redshifts outside of the ROI that have best-fiiting redshifts that fall within the ROI.

```python
numContamInROI = np.sum(permutationDataFrameContam[redshiftIsInROI(permutationDataFrameContam.BEST_FIT_REDSHIFT)].REDSHIFT_WEIGHT)
numContam = sum(permutationDataFrameContam.REDSHIFT_WEIGHT)
contamFraction = numContamInROI/numContam
# normalize by the relative extents of the contamination region and the ROI
roiExtent = roiRedshiftRange[1] - roiRedshiftRange[0]
contamExtent = (contamRedshiftRange[1] - contamRedshiftRange[0]) - roiExtent
contamExtentFraction = contamExtent/roiExtent

normContamFraction = contamFraction/contamExtentFraction

print (numContamInROI, numContam, contamFraction, roiExtent, contamExtent, contamExtentFraction, normContamFraction)
```

# Compute loss fraction

```python
numInROI = len(permutationDataFrameROI[redshiftIsInROI(permutationDataFrameROI.BEST_FIT_REDSHIFT)].index)
numOutsideROI = len(permutationDataFrameROI.index) - numInROI
lossFraction = numOutsideROI/(numInROI + numOutsideROI)
# normalize by the relative extents of the contamination region and the ROI
roiExtent = roiRedshiftRange[1] - roiRedshiftRange[0]
nonRoiExtent = (redshiftRange[1] - redshiftRange[0]) - (contamRedshiftRange[1] - contamRedshiftRange[0]) + roiExtent
lossExtentFraction = nonRoiExtent/roiExtent

normLossFraction = lossFraction/lossExtentFraction

print (numInROI, numOutsideROI, lossFraction, roiExtent, nonRoiExtent, lossExtentFraction, normLossFraction)
```
