# Compute and plot difference between true and best fit redshift

```python
permutationSubsetDataFrame = permutationDataFrame.iloc[0:len(dataSeries)].copy()
permutationSubsetDataFrame['BEST_FIT_REDSHIFT'] = dataSeries
permutationSubsetDataFrame['DELTA'] = (permutationSubsetDataFrame.BEST_FIT_REDSHIFT - permutationSubsetDataFrame.REDSHIFT)/(1 + permutationSubsetDataFrame.REDSHIFT)
deltaAxes = permutationSubsetDataFrame.DELTA.plot.hist(bins=100, histtype='step', fill=True, alpha=0.2)
deltaAxes.set_xlabel(r'$\frac{z_{\rm fit} - z_{\rm true}}{1+z_{\rm true}}$', size=15)
deltaDataCut = (permutationSubsetDataFrame.DELTA < 0.1) & (permutationSubsetDataFrame.DELTA > 0) & (permutationSubsetDataFrame.AGE > 3.5)
#display(permutationSubsetDataFrame[deltaDataCut])
#print(permutationSubsetDataFrame[deltaDataCut].index)

#display(permutationSubsetDataFrame.iloc[1060])
```

# Functions to plot grids of fit results grouped according to input model parameters

```python
def genGroupedRedshiftFitSubplots(data, groupbyCol, redshifts, groupByColUnit = None):
    grouped = data.groupby(groupbyCol)
    numGroups = grouped.ngroups
    # make subplots with three plots per row and as many rows as required
    numRows = np.ceil(numGroups/3)
    figure, axes = mplplot.subplots(figsize=(9,3*numRows), 
                                    nrows=int(numRows), 
                                    ncols=3)
    for plotIndex, (groupKey, plotAxis) in enumerate(zip(sorted(grouped.groups.keys()), axes.flatten())) :
        grouped.get_group(groupKey).BEST_FIT_REDSHIFT.plot.hist(ax=plotAxis, histtype='step', fill=True, alpha=0.2, bins=50, range=(0, 5))
        plotAxis.set_title('{} = {:4.2f} {}'.format(groupbyCol, groupKey, groupByColUnit if groupByColUnit is not None else ''))
        plotAxis.set_xlabel('Best Fit Redshift')
        plotAxis.vlines(redshifts, ymin=plotAxis.get_ylim()[0], ymax=plotAxis.get_ylim()[1], colors='r')

    mplplot.suptitle(groupbyCol, y=1.01)
    mplplot.tight_layout()
    
def genGroupedRedshiftFitDeltaSubplots(data, groupbyCol, redshifts, groupByColUnit = None):
    grouped = data.groupby(groupbyCol)
    numGroups = grouped.ngroups
    # make subplots with three plots per row and as many rows as required
    numRows = np.ceil(numGroups/3)
    figure, axes = mplplot.subplots(figsize=(9,3*numRows), 
                                    nrows=int(numRows), 
                                    ncols=3)
    for plotIndex, (groupKey, plotAxis) in enumerate(zip(sorted(grouped.groups.keys()), axes.flatten())) :
        grouped.get_group(groupKey).DELTA.plot.hist(ax=plotAxis, histtype='step', fill=True, alpha=0.2, bins=50)
        plotAxis.set_title('{} = {:4.2f} {}'.format(groupbyCol, groupKey, groupByColUnit if groupByColUnit is not None else ''))
        plotAxis.set_xlabel(r'$\frac{z_{\rm fit} - z_{\rm true}}{1+z_{\rm true}}$', size=15)
        
    mplplot.suptitle(groupbyCol, y=1.01)
    mplplot.tight_layout()
```

# Plots of summary statistics of difference ($z_{\rm fit}-z_{\rm true}$) versus SP AGE and $z_{\rm true}$

```python
deltaStatsFigure, deltaStatsAxes = mplplot.subplots(figsize=(12, 4),
                                                   ncols=3,
                                                   nrows=1)
for axisIndex, (function, title, axis) in enumerate(zip([np.mean, np.median, np.max], ['MEAN', 'MEDIAN', 'MAX'], deltaStatsAxes.flatten())):
    axes = permutationSubsetDataFrame.plot.hexbin(ax=axis, x='REDSHIFT', y='AGE', C='DELTA', gridsize=10, reduce_C_function=function, cmap='viridis', sharex=False)
    axes.set_title('{} DELTA'.format(title))
    axes.set_xlabel(r'$z_{\rm true}$', size=15)
    axes.set_ylabel('SP Age (Gyr)')

mplplot.tight_layout()
```
