```python
permutationSubsetDataFrame = permutationDataFrame[0:len(dataSeries)].copy()
permutationSubsetDataFrame['BEST_FIT_REDSHIFT'] = dataSeries
permutationSubsetDataFrame['DELTA'] = permutationSubsetDataFrame.BEST_FIT_REDSHIFT - permutationSubsetDataFrame.REDSHIFT
deltaAxes = permutationSubsetDataFrame.DELTA.plot.hist(bins=100, histtype='step', fill=True, alpha=0.2)
deltaAxes.set_xlabel('$z_{fit} - z_{true}$', size=15)
display(permutationSubsetDataFrame[permutationSubsetDataFrame.DELTA > 3])
print(permutationSubsetDataFrame[permutationSubsetDataFrame.DELTA > 3].index)
```

```python
def genGroupedSubplots(data, groupbyCol, redshifts, groupByColUnit = None):
    grouped = data.groupby(groupbyCol)
    numGroups = grouped.ngroups
    # make subplots with three plots per row and as many rows as required
    numRows = np.ceil(numGroups/3)
    figure, axes = mplplot.subplots(figsize=(9,3*numRows), 
                                    nrows=int(numRows), 
                                    ncols=3)
    for plotIndex, (groupKey, plotAxis) in enumerate(zip(sorted(grouped.groups.keys()), axes.flatten())) :
        grouped.get_group(groupKey).BEST_FIT_REDSHIFT.plot.hist(ax=plotAxis, histtype='step', fill=True, alpha=0.2, bins=50)
        plotAxis.set_title('{} = {:4.2f} {}'.format(groupbyCol, groupKey, groupByColUnit if groupByColUnit is not None else ''))
        plotAxis.set_xlabel('Best Fit Redshift')
        plotAxis.vlines(redshifts, ymin=plotAxis.get_ylim()[0], ymax=plotAxis.get_ylim()[1], colors='r')

    mplplot.suptitle(groupbyCol, y=1.01)
    mplplot.tight_layout()
```

```python
genGroupedSubplots(permutationSubsetDataFrame, "AGE", np.unique(permutationDataFrame.REDSHIFT.values), "Gyr")
```

```python
genGroupedSubplots(permutationSubsetDataFrame, "TAU", np.unique(permutationDataFrame.REDSHIFT.values), 'Gyr')
```

```python
genGroupedSubplots(permutationSubsetDataFrame, "METALLICITY_INDEX", np.unique(permutationDataFrame.REDSHIFT.values))
```
